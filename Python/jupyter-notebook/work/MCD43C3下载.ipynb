```python
import geemap
import ee
```

```python
Map = geemap.Map() #center参数是打开地图的中心点经纬度坐标
Map
```

```python
roi = geemap.geojson_to_ee("E:/Data/GIS/JX.json")
Map.centerObject(roi)
Map.addLayer(roi, {}, "roi")
```

```python
start_time = '2020-01' # 开始时间
end_time = '2021-01'   # 结束时间
```

```python
roi = Map.draw_last_feature.geometry()
```

```python
outpath = r'E:/Data/temp/'
def dow(i): # 用地图id做循环
    Layer = ee.Image(i).select('Albedo_BSA_shortwave') # 使用地图，select为确定地图波段，可以去除，则为全波段
    # 保存文件名为地图id加文件格式
    a = i.split('/')
    str = '-'
    b = str.join(a)
    file = outpath + b + '.tif'
    # 导出栅格影像，file_per_band为是否按波段分别导出
    geemap.ee_export_image(Layer, file, region=roi, scale=500)
```

```python
images = ee.ImageCollection("MODIS/006/MCD43C3")\
            .filterDate(start_time, end_time)
```

```python
name = images.aggregate_array('system:id').getInfo()   # 得到地图id
name
```

```python
from concurrent.futures.thread import ThreadPoolExecutor
with ThreadPoolExecutor(max_workers=5) as executor:
    executor.map(dow, name)
```

```python
for i in name:
    dow(i)
```

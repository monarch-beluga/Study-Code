```python
import geemap
import ee
```

```python
Map = geemap.Map()
Map
```

```python
file = r"E:/Paper/Data/roi_shp/jiujiang.shp"
points = geemap.shp_to_ee(file)
Map.addLayer(points, {}, "p")
Map.centerObject(points, 4)
```

```python
p1 = ee.Feature(points.first())
r1 = p1.buffer(500000)
t1 = r1.bounds()
```

```python
Map.addLayer(r1, {}, 'r1')
Map.addLayer(t1, {}, 't1')
```

```python
def fun(img):
    return img.select("B2", 'B3', 'B4')
```

```python
image = ee.ImageCollection("COPERNICUS/S2") \
    .filterBounds(points.geometry())\
    .filterDate('2020-06-21', '2020-06-24')
#     .map(fun)\
#     .filter(ee.Filter.calendarRange(160, 240, 'day_of_year'))\
#     .filterMetadata('CLOUDY_PIXEL_PERCENTAGE', 'less_than', 5)\
#     .sort('CLOUDY_PIXEL_PERCENTAGE', False)
```

```python
names = image.aggregate_array("system:id").getInfo()
```

```python
names
```

```python
i = []
ns = []
for name in names:
    if name[-6:] not in i:
        i.append(name[-6:])
        ns.append(name)
ns
```

```python
len(ns)
```

```python
imgs = []
for i in ns:
    imgs.append(ee.Image(i))
images = ee.ImageCollection(imgs).map(fun)
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
config = {
    "font.family": 'Times New Roman',
    "font.size": 14,
    "font.serif": 'Simsun',
}
matplotlib.rcParams.update(config)
```

```python
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = X ** 2 - 0.5 * X + np.random.randn(m, 1)
```

```python
plt.plot(X, y, 'b.')
plt.xlabel('X')
plt.ylabel('y')
plt.axis([-5, 5, -2, 13])
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)

X_poly = poly_features.fit_transform(X)
print(X[0], X_poly[0])
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
X_new = np.linspace(-3, 3, 100).reshape(-1, 1)
X_new_poly = poly_features.transform(X_new)
y_predict = lin_reg.predict(X_new_poly)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(X, y, 'b.')
plt.plot(X_new, y_predict, 'r--')
plt.xlabel('X')
plt.ylabel('y')
plt.axis([-5, 5, -2, 13])
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

plt.figure(figsize=(12, 6))
for color, degree in [['g--', 20], ['r--', 2], ['b--', 1]]:
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    std = StandardScaler()
    lin_reg = LinearRegression()
    poly_reg = Pipeline([
        ('poly_features', poly_features),
        ('std', std), ('lin_reg', lin_reg)])
    poly_reg.fit(X, y)
    y_new = poly_reg.predict(X_new)
    plt.plot(X_new, y_new, color, label=f'degree-{degree}', linewidth=1)
plt.plot(X, y, 'k.')
plt.xlabel('X')
plt.ylabel('y')
plt.axis([-5, 5, -2, 13])
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Ridge, Lasso

plt.figure(figsize=(12, 6))
lin_reg = LinearRegression()
lin_ridge = Ridge(alpha=0.1)
lin_lasso = Lasso(alpha=0.1)
colors = ['y--', 'g--', 'r--', 'b--']
degrees = [2, 20, 20 , 20]
models = [lin_reg, lin_reg, lin_ridge, lin_lasso]
labels = ["Linear-d2", "Linear-d20", "Ridge", "Lasso"]

for color, degree, model, label in zip(colors, degrees, models, labels):
    poly_features = PolynomialFeatures(degree=degree, include_bias=False)
    std = StandardScaler()
    
    poly_reg = Pipeline([
        ('poly_features', poly_features),
        ('std', std), ('lin_model', model)])
    poly_reg.fit(X, y)
    y_new = poly_reg.predict(X_new)
    plt.plot(X_new, y_new, color, label=label, linewidth=1)

    
plt.plot(X, y, 'k.')
plt.xlabel('X')
plt.ylabel('y')
plt.axis([-5, 5, -2, 13])
plt.legend()
plt.show()
```

### 样点合并

```python
from glob import glob
landuse_files = glob("shp/所有地类/*.shp")
```

```python
land_class = ['小麦', '建筑', '水体', '荒漠', '其他植被']
```

```python
import geopandas as gpd
landuse_shps = []
for file in landuse_files:
    gdf = gpd.read_file(file)
    gdf['TDLYDM'] = gdf['CID'].astype(int)
    gdf['TDLYMC'] = land_class[gdf['TDLYDM'][0]]
    gdf['CID'] = gdf.index
    landuse_shps.append(gdf[['CID', 'TDLYDM', 'TDLYMC', 'geometry']])
```

```python
import pandas as pd

landuse_shp = pd.concat(landuse_shps)
```

```python
landuse_shp.to_file("shp/landuse_point.shp", encoding='utf-8')
```

```python
import pandas as pd

label = 'TDLYDM'
def sample_points_to_df(img, ps, col_names=None):
    ps_data = img.sampleRegions(ps,
                                properties = [label],
                                geometries=True, 
                                scale=10)
    values_list = [f['properties'] for f in ps_data.getInfo()["features"]]
    if col_names == None:
        values_df = pd.DataFrame(values_list)
    else:
        values_df = pd.DataFrame(values_list)[col_names]
    return values_df
```

```python
from Monarch import user_gee
from scipy.signal import savgol_filter
import numpy as np

year = 2022
s2_year = []
lswis = []
ndvis = []
evis = []
times = []
for i in range(9, 21):
    year_s = year + i // 13
    year_e = year + (i+1) // 13
    month_s = i % 13 + i // 13
    month_e = (i+1) % 13 + (i+1) // 13
    s2 = ee.ImageCollection("COPERNICUS/S2_HARMONIZED")\
            .filterBounds(roi)\
            .sort('CLOUD_COVERAGE_ASSESSMENT', False)\
            .filterDate(f"{year_s}-{month_s}", f"{year_e}-{month_e}")
    s2_rm_cloud = s2.map(user_gee.rm_cloud_s2_sr)
    s2_z = s2_rm_cloud.merge(s2).mosaic().select(['B2', 'B3', 'B4', 'B5', 'B8', 'B11'])
    a = s2.aggregate_array('system:time_start').map(lambda i: ee.Date(i).format('YYYY-MM-dd'))
    evi = s2_z.expression(
        '2.5 * ((NIR - RED) / (NIR + 6 * RED - 7.5 * BLUE + 1))', {
        'NIR': s2_z.select('B8'),
        'RED': s2_z.select('B4'),
        'BLUE': s2_z.select('B2')
    }).rename(f"S2_EVI")
    ndvi = s2_z.normalizedDifference(['B8', 'B4']).rename(f"S2_NDVI")
    evis.append(evi)
    ndvis.append(ndvi)
    s2_image = s2_z.addBands(ndvi)
    s2_image = s2_image.addBands(evi)
    s2_year.append(s2_image)
    times.append(f"{year_s}.{month_s}")

index_names = ["NDVI", "EVI"]
index_imgs = [ndvis, evis]
data_methods = ["mean", "median"]
for index_name, index_img in zip(index_names, index_imgs):
    sample_ps_txt = f"txt_data/{index_name}_sample_ps_{year}.txt"
    if os.path.exists(sample_ps_txt):
        index_df = pd.read_csv(sample_ps_txt)[[f'{i}_S2_{index_name}' for i in range(12)]]
    else:
        index_image = ee.ImageCollection(index_img).toBands().clip(roi)
        index_df = sample_points_to_df(index_image, wheat_ps)[[f'{i}_S2_{index_name}' for i in range(12)]]
        index_df.to_csv(sample_ps_txt)
    index_data = index_df.values
    index_mean = np.mean(index_data, axis=0)
    index_median = np.median(index_data, axis=0)
    index_mean_sg = savgol_filter(index_mean, 7, 3)
    index_median_sg = savgol_filter(index_median, 7, 3)
    for data_method, data in zip(data_methods, [index_mean_sg, index_median_sg]):
        fig, ax = plt.subplots(figsize=(12, 8))
        ax.set_title(f"{index_name} {data_method} change in {year}", weight="bold", size=18, family='Times New Roman')
        ax.set_xlabel('month', family='Times New Roman')
        ax.set_ylabel(f'{index_name}', family='Times New Roman')
        ax.tick_params("both", direction='in')
        labels = ax.get_xticklabels() + ax.get_yticklabels()
        [label.set_fontname('Times New Roman') for label in labels]
        ax.plot(times, data, '-o', color='c', markersize=3, lw=1, label=f'{index_name}')
        ax.legend(frameon=False, handlelength=2)
        fig.savefig(f'png_data/{index_name}_{data_method}_sg_{year}.png', dpi=600, bbox_inches='tight')
```

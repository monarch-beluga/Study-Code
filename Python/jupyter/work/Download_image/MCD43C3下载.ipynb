```python
import geemap
import ee
import os
from concurrent.futures.thread import ThreadPoolExecutor
```

```python
Map = geemap.Map() #center参数是打开地图的中心点经纬度坐标
Map
```

```python
roi = geemap.shp_to_ee(r'E:/Data/clip/roi.shp')
Map.centerObject(roi)
Map.addLayer(roi, {}, "roi")
```

```python
path = r'D:/Albedo/'
os.chdir(path)
```

```python
for year in range(2015, 2016):
    if year == 2018:
        continue
    images = ee.ImageCollection("MODIS/006/MCD43C3")\
            .filterDate(str(year), str(year+1))

    names = images.aggregate_array('system:id').getInfo()
    
    def dow(i):
        start = i
        end = i+8
        if end > len(names):
            end = len(names)
        imgs = [ee.Image(names[j]) for j in range(start, end)]
        img = ee.ImageCollection(imgs).select('Albedo_BSA_shortwave')\
            .map(lambda i: i.clip(roi.geometry())).mean()
        name = names[i].split('/')[-1]
        geemap.ee_export_image(img, r'BAS_{0}{1:03d}.tif'.format(year,i+1), 
                               scale=1000, region=roi.geometry())
    
    length = images.size().getInfo()
    List = list(range(0, length, 8))
    with ThreadPoolExecutor(max_workers=10) as executor:
        executor.map(dow, List)
```

```python
i = 0
imgs = [ee.Image(names[j]) for j in range(i, i+8)]
img = ee.ImageCollection(imgs).select('Albedo_BSA_shortwave')\
            .map(lambda i: i.clip(roi.geometry())).mean()
Map.addLayer(img, {'min': 0, 'max': 400}, 't')
```

```python
outpath = r'F:/Albedo/'
def dow(i): # 用地图id做循环
    Layer = ee.Image(i).select('Albedo_BSA_shortwave') # 使用地图，select为确定地图波段，可以去除，则为全波段
    # 保存文件名为地图id加文件格式
    a = i.split('/')
    str = '-'
    b = str.join(a)
    file = outpath + b + '.tif'
    # 导出栅格影像，file_per_band为是否按波段分别导出
    geemap.ee_export_image(Layer, file, region=roi, scale=500)
```
